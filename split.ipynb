```python
import os
import numpy as np
from keras.preprocessing import image
import matplotlib.pyplot as plt
%matplotlib inline

def load_and_save_images_from_path(path, save_to_folder, resize_to=(1024, 1024), crop_to=(128, 128)):
    if not os.path.exists(save_to_folder):
        os.makedirs(save_to_folder)  # Create folder if it doesn't exist
    
    images = []
    j = 1
    for file in os.listdir(path):
        full_path = os.path.join(path, file)
        if os.path.isfile(full_path):  # Check if it's a file
            # Step 1: Load the image and resize to 1024x1024
            img = image.load_img(full_path, target_size=resize_to)
            img_array = image.img_to_array(img)
            for i in range(4) :
                # Step 2: Crop the resized image to 128x128
                height, width, _ = img_array.shape
                crop_h, crop_w = crop_to
                start_h = (height - crop_h + randomize[os.listdir(path).index(file) + i]) // 2
                start_w = (width - crop_w + randomize[os.listdir(path).index(file) + i]) // 2
                cropped_img = img_array[start_h:start_h+crop_h, start_w:start_w+crop_w, :]
                # Save the cropped image
                cropped_img_name = "image_cropped" + "_" + str(j) + ".png"
                j += 1
                save_path = os.path.join(save_to_folder, cropped_img_name)
                image.save_img(save_path, cropped_img)  # Save the cropped image
                images.append(cropped_img)
    return np.array(images)  # Convert to NumPy array

def show_images(images):
    # Show up to 8 images
    fig, axes = plt.subplots(1, min(len(images), 8), figsize=(16, 16), subplot_kw={'xticks': [], 'yticks': []})

    for i, ax in enumerate(axes.flat[:min(len(images), 8)]):
        ax.imshow(images[i].astype("uint8"))
```

```python
# Initialize x_train as a NumPy array from the start
x_train = np.empty((0, 128, 128, 3))  # Shape for cropped images
y_train = np.empty((0, 128, 128, 3))

# Define save directories
input_save_dir = 'PCB_Image_final/Open_circuit_bin_resized'
output_save_dir = 'PCB_Image_final/Open_circuit_bin_output_resized'

# Load, resize, crop, and save the training images
images = load_and_save_images_from_path('PCB_Image_final/train/Open_circuit_bin', input_save_dir, resize_to=(1024, 1024), crop_to=(128, 128))

if images.size > 0:  # Ensure images are loaded
    show_images(images)  # Display the cropped images
    x_train = np.concatenate((x_train, images), axis=0)  # Concatenate images to x_train

# Load, resize, crop, and save the label images
images = load_and_save_images_from_path('PCB_Image_final/train/Open_circuit_bin_output', output_save_dir, resize_to=(1024, 1024), crop_to=(128, 128))

if images.size > 0:  # Ensure images are loaded
    show_images(images)  # Display the cropped images
    y_train = np.concatenate((y_train, images), axis=0)  # Concatenate images to y_train

# Normalize the image arrays
x_train = x_train / 255.0
y_train = y_train / 255.0

print(f"Shape of cropped input images: {x_train.shape}")
print(f"Shape of cropped output images: {y_train.shape}")
```
